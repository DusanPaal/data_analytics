### 2. Data sampling, model compilation and training

```python
batch_size = 20
img_width = 150
img_height = 150

assert img_width == img_height, 'The image shape must be a square!'

# data sampling
train_generator = image_dataset_from_directory(
    os.path.join('data', 'dogs_cats', 'train'),
    image_size=(img_width, img_height),
    batch_size=batch_size,
    label_mode='binary'
).repeat()  # Ensures dataset repetition every epoch

validation_generator = image_dataset_from_directory(
    os.path.join('data', 'dogs_cats', 'validation'),
    image_size=(img_width, img_height),
    batch_size=batch_size,
    label_mode='binary'
).repeat()  # Ensures dataset repetition every epoch

test_generator = image_dataset_from_directory(
    os.path.join('data', 'dogs_cats', 'test'),
    image_size=(img_width, img_height),
    batch_size=batch_size,
    label_mode='binary'
).repeat()  # Ensures dataset repetition every epoch

# Define image augmentation layers
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),        # Random horizontal flip
    layers.RandomRotation(0.4),             # Randomly rotate images by 20%
    layers.RandomZoom(0.2),                 # Randomly zoom in/out by 20%
    layers.RandomTranslation(0.2, 0.2),     # Randomly translate images
    layers.RandomContrast(0.2)              # Randomly change contrast
])

# Apply the data augmentation to the training dataset
train_dataset = train_generator.map(
    # Apply augmentations only during training
    lambda x, y: (data_augmentation(x, training=True), y),  
    num_parallel_calls=tf.data.AUTOTUNE
)

print('----------------------------------------')

# initilaize new CNN model
model_ref = models.Sequential()

# input layer for RGB images, hence 3 channels
model_ref.add(layers.Input(shape=(img_width, img_height, 3))) 

# Block 1: recognizes small local patterns
model_ref.add(layers.Conv2D(64, (3, 3), activation='relu'))
model_ref.add(layers.Conv2D(64, (3, 3), activation='relu'))
model_ref.add(layers.MaxPooling2D((2, 2)))

# Block 1: recognizes small local patterns
model_ref.add(layers.Conv2D(128, (3, 3), activation='relu')) 
model_ref.add(layers.Conv2D(128, (3, 3), activation='relu')) 
model_ref.add(layers.MaxPooling2D((2, 2)))
model_ref.add(layers.Dropout(0.35))

# Block 2: recognizes larger patterns from prev layer
model_ref.add(layers.Conv2D(256, (3, 3), activation='relu'))
model_ref.add(layers.Conv2D(256, (3, 3), activation='relu')) 
model_ref.add(layers.MaxPooling2D((2, 2)))
model_ref.add(layers.Dropout(0.35))

# Block 2: recognizes larger patterns from prev layer
model_ref.add(layers.Conv2D(256, (3, 3), activation='relu')) 
model_ref.add(layers.Conv2D(256, (3, 3), activation='relu')) 
model_ref.add(layers.MaxPooling2D((2, 2)))
# model_ref.add(layers.BatchNormalization())
model_ref.add(layers.Dropout(0.4))

# now adding dense layers for classification;
# dense layers work with 1D tensors so flatten
# the output (3D feature maps) of the last layer
model_ref.add(layers.Flatten())
model_ref.add(layers.Dense(
    256, activation='relu',
    # every coefficient w in the W matrix of 
    # a layer adds 0.05 * w to the total loss
    kernel_regularizer=regularizers.l2(0.08)) 
) 
model_ref.add(layers.Dropout(0.5))
model_ref.add(layers.Dense(1, activation='sigmoid'))

# compile the model 
model_ref.compile(
    optimizer=optimizers.Adam(2e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# prevent validation accuracy to drop
# when training takes too long
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    restore_best_weights=True
)

# optimisations that prevent fluctuations
# in validation loss at the end of training
learning_rate_scheduler = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.2,
    patience=5,
    min_lr=1e-6
)

# train the model
history = model_ref.fit(
    train_dataset,
    steps_per_epoch=2000 // batch_size,
    batch_size=batch_size,
    epochs=150,
    validation_data=validation_generator,
    validation_steps=1000 // batch_size,
    # callbacks=[learning_rate_scheduler]
)

model_path = join('models', 'dogs_cats', 'reference.keras')
model_ref.save(model_path)
print('Model saved to:', model_path)

# evaluate the model in terms training and validation loss
test_loss, test_acc = model_ref.evaluate(test_generator, steps=50)
print('Test loss:', round(test_loss, 4), 'Test accuracy:', round(test_acc, 4))

history_loss = history.history['loss']
history_val_loss = history.history['val_loss']
epochs=range(1, len(history_loss) + 1)
plot_size = (14, 5)

plt.figure(figsize=plot_size)
plt.plot(epochs, history_loss, 'g', label='Training loss')
plt.plot(epochs, history_val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# evaluate the model in terms of accuracy and loss
history_accuracy = history.history['accuracy']
history_val_accuracy = history.history['val_accuracy']

plt.figure(figsize=plot_size)
plt.plot(epochs, history_accuracy, 'g', label='Training accuracy')
plt.plot(epochs, history_val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import keras
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
import numpy as np

conv_base = VGG16(    
    include_top=False,
    weights="imagenet",
    input_shape=(150, 150, 3)
)


datagen = ImageDataGenerator(rescale=1.0/255)
batch_size = 20

def extract_features(directory, n_samples):
    
    feats = np.zeros(shape=(n_samples, 4, 4, 512))
    labels = np.zeros(shape=(n_samples))
    
    generator = datagen.flow_from_directory(
        directory,
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='binary'        
    )
    i = 0

    for inputs_batch, labels_batch in generator:
        feats_batch = conv_base.predict(inputs_batch)
        feats[i * batch_size: (1 + i) * batch_size] = feats_batch
        labels[i * batch_size: (1 + i) * batch_size] = labels_batch
        i += 1
        if i >= batch_size:
            break

    return feats, labels

train_dir = os.path.join('data', 'dogs_cats', 'train')
valid_dir = os.path.join('data', 'dogs_cats', 'validation')
test_dir = os.path.join('data', 'dogs_cats', 'test')
train_feats, train_labels = extract_features(train_dir, 2000)
valid_feats, valid_labels = extract_features(valid_dir, 1000)
test_feats, test_labels = extract_features(test_dir, 1000)

# flatten the feats before using in a custom dense classifier
train_feats = np.reshape(train_feats, (2000, 4*4*512))
valid_feats = np.reshape(valid_feats, (1000, 4*4*512))
test_feats = np.reshape(test_feats, (1000, 4*4*512))

# crate the custom dense classifier
model = models.Sequential()
# model.add(layers.Input(shape=(4*4*512,)))
model.add(layers.Dense(256, activation='relu', input_dim=4*4*512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(
    optimizer=optimizers.RMSprop(learning_rate=2e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_feats, train_labels,
    epochs=30,
    batch_size=batch_size,
    validation_data=(
        valid_feats,
        valid_labels
    )
)

# evaluate the model in terms training and validation loss
# test_loss, test_acc = model.evaluate(test_feats, steps=50)
# print('Test loss:', round(test_loss, 4), 'Test accuracy:', round(test_acc, 4))

history_loss = history.history['loss']
history_val_loss = history.history['val_loss']
epochs=range(1, len(history_loss) + 1)
plot_size = (14, 5)

plt.figure(figsize=plot_size)
plt.plot(epochs, history_loss, 'g', label='Training loss')
plt.plot(epochs, history_val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# evaluate the model in terms of accuracy and loss
history_accuracy = history.history['accuracy']
history_val_accuracy = history.history['val_accuracy']

plt.figure(figsize=plot_size)
plt.plot(epochs, history_accuracy, 'g', label='Training accuracy')
plt.plot(epochs, history_val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Model
from glob import glob

def read_image(img_path: str, raw=False) -> np.ndarray:
    
    img = image.load_img(img_path, target_size=(150, 150))

    if raw:
        return img
        
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    img_tensor /= 255.0
    
    return img_tensor

model_path = join('models', 'dogs_cats', 'reference.keras')
model_loaded = models.load_model(model_path, compile=False)

d = {
    'image': [],
    'prediction': [],
    'category': [],
    'desc': []
}

import keras

for categ in ['dog', 'cat']:

    images = glob(f'data/dogs_cats/test/{categ}s/*.jpg')
    
    for img_path in images:
        img = read_image(img_path)
        result = model_loaded.predict(img, verbose=0)
        p = float(result[0][0])
        d['image'].append(img_path)
        d['prediction'].append(p)
        d['category'].append(categ)
        score = float(keras.ops.sigmoid(p))
        d['desc'].append(f"This image is {100 * (1 - score):.2f}% cat and {100 * score:.2f}% dog.")

predictions = pd.DataFrame.from_dict(d).sort_values('prediction').reset_index()
predictions.describe()
```
